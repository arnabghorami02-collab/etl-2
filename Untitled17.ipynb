1 : Define Data Transformation in ETL and explain why it is important.

Sol :   

-  Data Transformation in ETL is the process of converting raw data into a clean, organized, and usable format before loading it into a database or data warehouse.

It is important because:

- It improves data quality.
- It makes data consistent and standardized.
- It helps in accurate analysis and reporting.
- It converts data into a format suitable for business needs.


2 : List any four common activities involved in Data Cleaning.

Sol :       

 Four common activities involved in Data Cleaning are:

- Removing duplicate data
- Handling missing values
- Correcting inconsistent formats (e.g., Male/male/M)
- Removing invalid or incorrect data

3 : What is the difference between Normalization and Standardization?

Sol :       
- Normalization (e.g., Min-Max Scaling): Rescales the data values to fit within a specific bounded range, typically 0 to 1. It is highly influenced by outliers and is often used when the data does not follow a normal (Gaussian) distribution.
- Standardization (e.g., Z-score Scaling): Centers the data around a mean of 0 with a standard deviation of 1. It does not bound the data to a specific range and is much more robust to outliers. It is preferred when the data follows a normal distribution or when algorithms assume normal distributions.

7 : Explain the difference between Data Integration and Data Mapping in ETL.

Sol :      

- Data Integration is the overarching process of combining data from multiple, disparate sources into a single, unified view or storage destination.

- Data Mapping is a specific, foundational step within data integration. It is the process of defining the relationship between the fields in the source data and the fields in the target destination (e.g., establishing that the "cust_id" column in Source A matches the "CustomerID" column in Target B).

8 : Explain why Z-score Standardization is preferred over Min-Max Scaling when outliers exist.

Sol :        

Z-score Standardization is preferred over Min-Max Scaling when outliers exist because Min-Max Scaling depends heavily on minimum and maximum values. Extreme outliers can compress most data into a small range.


Z-score Standardization reduces the effect of outliers by using mean and standard deviation instead of min and max.
